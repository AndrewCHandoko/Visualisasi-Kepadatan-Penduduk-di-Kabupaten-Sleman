### Convert Data Type Float Menjadi Dict (List) Kembali

```python
def castToList(x): 
    if isinstance(x, list):
        return x
    elif isinstance(x, str):
        return [x]
    try:
        return list(x)
    except TypeError:
        return [x]

n = list(kepadatan_penduduk)
kepadatan_penduduk=castToList(n)
```

```python
print(castToList(n))
```

### Menambahkan Kolom Kepadatan Penduduk Ke Dalam File CSV  

```python
penduduk['kepadatan_penduduk'] = (n)
```

```python
penduduk
```

### Membuat Klasifikasi Kepadatan Penduduk 

```python
klasifikasi = list()

for x in range(0,17):
    if(penduduk['kepadatan_penduduk'][x] < 500):
        klasifikasi.append("1")
    elif(1249 < penduduk['kepadatan_penduduk'][x] > 500):
        klasifikasi.append("2")
    elif(2499 < penduduk['kepadatan_penduduk'][x] > 1250):
        klasifikasi.append("3")
    elif(3999 < penduduk['kepadatan_penduduk'][x] > 2500):
        klasifikasi.append("4")
    elif(5999 < penduduk['kepadatan_penduduk'][x] > 4000):
        klasifikasi.append("5")
    elif(7499 < penduduk['kepadatan_penduduk'][x] > 6000):
        klasifikasi.append("6")
    elif(8499 < penduduk['kepadatan_penduduk'][x] > 7500):
        klasifikasi.append("7")
    elif(penduduk['kepadatan_penduduk'][x] > 8500):
        klasifikasi.append("8")
    else:
        klasifikasi.append("9")
```

```python
klasifikasi
```

### Convert File Menjadi File GeoDataFrame & Memasukkan Kolom Baru Yaitu Klasifikasi  

```python
geo_penduduk=gpd.GeoDataFrame(penduduk)
```

```python
geo_penduduk['Klasifikasi']=klasifikasi
```

```python
geo_penduduk
```

### Mengganti Nama Kolom 'kecamatan' Menjadi 'kec_2008'

```python
geo_penduduk=geo_penduduk.rename(columns={'kecamatan':'kec_2008'})
```

```python
geo_penduduk
```

### Menggabungkan Tabel Data Pada File SHP Kab Sleman Dengan File CSV Kepadatan Penduduk
